```python
# -*- coding:UTF-8 -*-
import io
import json 

train_name = 'flickr30kzhbbosontrain/ImageSets/flickr30kzhbbosontrain.txt'
train_caption = 'flickr30kzhbbosontrain/TextData/seg.flickr30kzhbbosontrain.caption.txt'
val_name = 'flickr30kzhbbosonval/ImageSets/flickr30kzhbbosonval.txt'
val_caption = 'flickr30kzhbbosonval/TextData/seg.flickr30kzhbbosonval.caption.txt'
test_name = 'flickr30kzhmbosontest/ImageSets/flickr30kzhbbosontest.txt'
test_caption = 'flickr30kzhmbosontest/TextData/seg.flickr30kzhmbosontest.caption.txt'

dict1 = {} 
images = []
# creating dictionary 
with io.open(train_caption, 'r') as f: 
    content = f.readlines()
    f.close()
    old_filename = "1000092795"
    sentences = []
    for line in content: 
        #print(line[30:31])
        if line.find("#")==-1:continue
        new_filename, _, caption = line.strip().split("#",2)
        #print(new_filename)
        if new_filename != old_filename:
            #print("filename",old_filename, "split","train", "sentences",sentences)
            images.append({"filename":old_filename+".jpg", "split":"train", "sentences":sentences})
            sentences = []
        sentences.append({"caption_id":caption[0], "tokens":caption[2:]}) 
        #print(sentences)
        old_filename = new_filename
        if old_filename == "998845445":
            images.append({"filename":old_filename+".jpg", "split":"train", "sentences":sentences})
            
            
# creating dictionary 
with io.open(val_caption, 'r') as f: 
    content = f.readlines()
    f.close()
    old_filename = "104816788"
    sentences = []
    for line in content: 
        #print(line[30:31])
        if line.find("#")==-1:continue
        new_filename, _, caption = line.strip().split("#",2)
        #print(new_filename)
        if new_filename != old_filename:
            #print("filename",old_filename, "split","train", "sentences",sentences)
            images.append({"filename":old_filename+".jpg", "split":"val", "sentences":sentences})
            sentences = []
        sentences.append({"caption_id":caption[0], "tokens":caption[2:]}) 
        old_filename = new_filename
        if old_filename == "989754491":
            images.append({"filename":old_filename+".jpg", "split":"val", "sentences":sentences})
            
            
# creating dictionary 
with io.open(test_caption, 'r') as f: 
    content = f.readlines()
    f.close()
    old_filename = "1009692167"
    sentences = []
    for line in content: 
        #print(line[30:31])
        if line.find("#")==-1:continue
        new_filename, _, caption = line.strip().split("#",2)
        #print(new_filename)
        if new_filename != old_filename:
            #print("filename",old_filename, "split","train", "sentences",sentences)
            images.append({"filename":old_filename+".jpg", "split":"test", "sentences":sentences})
            sentences = []
        text=caption[2:].split()
        text=[x.split(":")[0] for x in text]
        sentences.append({"caption_id":caption[0], "tokens":str(" ".join(text))}) 
        old_filename = new_filename
        if old_filename == "977856234":
            images.append({"filename":old_filename+".jpg", "split":"test", "sentences":sentences})
# creating json file 
# the JSON file is named as test1 
print(len(images))
dict1["images"] = images
out_file = open("flickr30kzh.json", "w") 
json.dump(dict1, out_file, indent = 4, sort_keys = False) 
out_file.close() 
```
